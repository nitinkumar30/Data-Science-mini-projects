```python
# import required libraries
from bs4 import BeautifulSoup
# import urllib2 ; instead of this, execute following command
from urllib.request import urlopen
```

```python
# url for webscrapping
url = 'https://simplilearn.com/resources'
```

```python
# scrap the webpage document
# webpage = urllib2.urlopen(url)   #Change this step to below one
webpage = urlopen(url)
```

```python
# Create a soup object and parse the webpage using html parser
sl_soup = BeautifulSoup(webpage, 'html.parser')
```

```python
# close the webpage object
webpage.close()
```

```python
# view the content of the soup object
sl_soup.contents
```

```python
# use prettify method to view well formatted html output
print(sl_soup.prettify())
```

```python
# view the title of the webpage
sl_soup.title
```

```python
# find all href link present in the webpage
for href in sl_soup.findAll("a", href=True):
    print(href['href'])
```

```python
# view all the headers name
for article in sl_soup.findAll("h2"):
    print(article.string)
```

```python
for article_topic in sl_soup.findAll("h4"):
    print(article_topic.string)
```

```python
url2 = "https://www.simplilearn.com/what-is-agile-project-management-article?source=frs_home"
webpage2 = urlopen(url2)
sl_article = BeautifulSoup(webpage2, 'html.parser')
test = sl_article.find(class_="desig_author empty-text")
```

```python
# view the type of the article object
type(test)
```

```python
# find all article name
test.findAll('p')
```

```python
# find first article name
first_next_p = test.p
first_next_p
```

```python
# find next sibling (next article names)
find_next_sibling = first_next_p.next_sibling.next_sibling
find_next_sibling
```

```python
# find previous siblings (previous article names)
find_next_sibling.previous_sibling.previous_sibling
```
